```python
!pip install pyspark  # Install PySpark in Google Colab

from pyspark import SparkContext
from pyspark.sql import SparkSession

# Create a SparkContext
#sc = SparkContext("local[*]", "MyApp")

# Create a SparkSession from the SparkContext
spark = SparkSession.builder.appName("MyApp").getOrCreate()

# Read input file and split each line by comma
lines = spark.sparkContext.textFile("/content/access.csv")

# Perform further operations on the RDD or convert it to a DataFrame and use Spark SQL
ips = lines.map(lambda x: x.split(",")[0])

# Map each IP to (IP, 1) for counting
counts = ips.map(lambda x: (x, 1))

# Reduce by key to count occurrences of each IP
sums = counts.reduceByKey(lambda x, y: x + y)

# Sort IP addresses by count in descending order
sorted_ips = sums.sortBy(lambda x: x[1], ascending=False)

# Collect and print the output
output = sorted_ips.collect()
for ip, count in output:
    print(f"IP: {ip}, Count: {count}")
```
